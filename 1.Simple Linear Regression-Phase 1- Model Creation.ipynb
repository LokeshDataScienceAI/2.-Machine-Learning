```python
import pandas as pd #Pandas is a library that handles single to multiple datasets; therefore, it is commonly imported as pd
```

```python
dataset=pd.read_csv("Salary_Data.csv")# pd is an alias for the pandas library. The read_csv() function is called using pd, and the parentheses () are used to pass parameters to the function
```

```python
dataset #see the data set
```

```python
independent=dataset[["YearsExperience"]] #independent means input values therefore independent from dataset values
independent
```

```python
dependent=dataset[["Salary"]] #dependent means output values therefore dependent from dataset values
dependent
```

```python
from sklearn.model_selection import train_test_split #sklearn.model is a library and this command all algorithim main concept
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0) #x-input value, y-output value, so split the x and y train and test
```

```python
from sklearn.linear_model import LinearRegression# see the linear regression value from sklearn model
regressor=LinearRegression()#run the regressor 
regressor.fit(x_train,y_train)
```

```python
weight=regressor.coef_ #check the weight value
weight
```

```python
bais=regressor.intercept_ #check the bais value and it is 0 value that means initial starting value of data set =26777
bais
```

```python
y_pred=regressor.predict(x_test) #predicition the regression value
```

```python
from sklearn.metrics import r2_score #imported the r2_score from sklearn.metrics
r_score=r2_score(y_test,y_pred)#check the r_score value
```

```python
r_score #as per rule nearly 1 percentage is good value 
```

```python
import pickle #pickle is liberary for use save the file name
filename="Finalized_model_linear.sav" #if use save the file name
```

```python
pickle.dump(regressor,open(filename,'wb')) #if it is used for write a binary
```

```python
loaded_model=pickle.load(open("Finalized_model_linear.sav",'rb')) #load a model with read the binary in save file name
result=loaded_model.predict([[13]]) #check the value for experience
```
